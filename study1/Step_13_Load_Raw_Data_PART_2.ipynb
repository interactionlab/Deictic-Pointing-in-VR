```python
import numpy as np
import pandas as pd

OUTPUT_PATH = "./out/"
```

```python
%%time
dfBone = pd.read_pickle(OUTPUT_PATH + "data_bone_v01.pkl")
dfBone= dfBone.sort_values("TimeUnix")
dfBone.head()
```

```python
%%time
dfRawPointer = pd.read_pickle(OUTPUT_PATH + "data_raw_Pointer_v01.pkl")
dfRawPointerFILTER = dfRawPointer[['SessionId', 'PId', 'TimeUnix', 'IsPointer', 'RightLowerArmRot', 'RightLowerArmPos']]
dfRawPointerFILTER.head()
```

```python
%%time 
dfMatch = pd.merge_asof(dfBone, dfRawPointerFILTER, on='TimeUnix', by=['SessionId', 'PId', 'IsPointer'])
dfMatch.head()
```

```python
%%time
dfMatch.to_pickle(OUTPUT_PATH + "data_bone_v02.pkl")
```

```python
dfBone = dfMatch
```

```python
%%time
dfClick = pd.read_pickle(OUTPUT_PATH + "data_click_v01.pkl")
dfClick.head()
```

```python
len(dfClick[dfClick.IsPointer == True])
```

```python

len(dfClick[dfClick.IsPointer == False])
```

```python
%%time 
dfBonePointer = dfBone[(dfBone.IsPointer == True)].sort_values("Time")
dfBoneObserver = dfBone[(dfBone.IsPointer == False)].sort_values("Time")

dfC = dfClick[(dfClick.IsPointer == False)].sort_values("Time")
dfs = []
for i, dfR in enumerate ([dfBonePointer, dfBoneObserver]):
    dfMatch = pd.merge_asof(dfC, dfR, on="Time", by="SessionId")
    dfMatch.rename(columns={'PId_y': 'PId', 'TimeUnix_x': 'TimeUnixClick', 'TimeUnix_y': 'TimeUnixBone',
                            'StartTimeUnix_x': 'StartTimeUnixClick', 'StartTimeUnix_y': 'StartTimeUnixBone',
                            'IsPointer_y': 'IsPointer', 'File_x': 'FileClick', 'File_y': 'FileBone'}, inplace=True)
    del dfMatch["PId_x"]
    del dfMatch["IsPointer_x"]
    dfs.append(dfMatch)
dfMatch = pd.concat(dfs)
```

```python
dfMatch.groupby(["PId", "IsPointer"]).count()
```

```python
dfMatch = dfMatch.sort_values("Time")
dfMatch.index = list(range(len(dfMatch)))
dfMatch.to_pickle(OUTPUT_PATH + "data_v02.pkl")
```
